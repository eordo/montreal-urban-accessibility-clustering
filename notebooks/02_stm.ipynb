# 2. STM

Although I do not model travel by transit nor account for intermodal travel, e.g., walking then taking a bus, I count transit stops as a special POI in the spirit of the 15-minute city.
The list of metro stations, bus stations, and bus stops are provided by the [_Société de transport de Montréal (STM)_ (Montreal Transit Corporation)](https://www.stm.info/en/about/developers).

```python
from final_project.config import FEATURES_DIR, RAW_DIR
from final_project.data import stm
```

The STM operates throughout Greater Montreal, so the provided list is restricted to stops and stations only on the Island of Montreal.
Ensure the following shapefiles are extracted to the data directory with the required names.

```python
island_shp = RAW_DIR / 'island' / 'limites-terrestres.shp'
stm_shp = RAW_DIR / 'stm' / 'stm_arrets_sig.shp'

stops = stm.load_transit_stops(stm_shp, island_shp)
stops.head()
```

Save this data for feature engineering in the next steps.

```python
stops.to_file(FEATURES_DIR / 'stops.geojson', driver='GeoJSON')
```
